```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
import os
```

```python
path = kagglehub.dataset_download("muhammadsajad/concertriccir2")

print("Path to dataset files:", path)
print(os.listdir(path))
```

```python
df = pd.read_csv('/root/.cache/kagglehub/datasets/muhammadsajad/concertriccir2/versions/1/concertriccir2.csv')
df.head()
```

```python
plt.scatter(df['7.003345706103683010e-01'], df['-2.470675778972781789e-01'],c=df['0.000000000000000000e+00'])
```

```python
x = df.drop('0.000000000000000000e+00',axis = 1)
y = df['0.000000000000000000e+00']
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()
model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
history1 = model.fit(x,y,epochs=200,validation_split=0.2)
```

```python
model = Sequential()
model.add(Dense(3,activation='relu',input_dim=2))
model.add(BatchNormalization())
model.add(Dense(2,activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
history2 = model.fit(x,y,epochs=200,validation_split=0.2)
```

```python
plt.plot(history1.history['loss'],color='black')
plt.plot(history2.history['loss'],color='green')
```
